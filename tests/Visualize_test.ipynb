```python
from tf.app import use

mt = use('etcbc/bhsa')
mtF, mtL, mtT = mt.api.F, mt.api.L, mt.api.T
sp = use('DT-UCPH/SamP:clone', checkout='clone', version='1.3')
spF, spL, spT  = sp.api.F, sp.api.L, sp.api.T
```

```python
sp_book = set([spT.sectionFromNode(w)[0] for w in spF.otype.s('word')])
sp_lexemes = [spF.lex.v(w) for w in spF.otype.s('word')]
mt_lexemes = [mtF.lex.v(w) for w in mtF.otype.s('word') if mtT.sectionFromNode(w)[0] in sp_book]
```

```python
unique_sp_lexemes = set()
unique_mt_lexemes = set()

for w in sp_lexemes:
    if w not in mt_lexemes:
        unique_sp_lexemes.add(w)
        
for w in mt_lexemes:
    if w not in sp_lexemes:
        unique_mt_lexemes.add(w)
        
print(len(unique_sp_lexemes))
print(len(unique_mt_lexemes))
```

```python
check_lexemes = list(unique_mt_lexemes)
check_lexemes.sort()
```

#### MT lexemes with variation between D and R

```python
import collections

g_cons_dict = collections.defaultdict(set)

for l in mtF.otype.s('lex'):
    if not ('H' in mtF.lex.v(l) and 'X' in mtF.lex.v(l)):
        for w in mtL.d(l, 'word'):
            g_cons_dict[mtF.lex.v(l)].add(mtF.g_cons.v(w))
            
for l in g_cons_dict:
    D = False
    R = False
    for w in g_cons_dict[l]:
        if 'H' in w:
            D = True
        if 'X' in w:
            R = True
    if D and R:
        print(l)
```
