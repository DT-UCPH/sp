# Test morphology

```python
from tf.app import use

MT = use('etcbc/bhsa', hoist=globals())
SP = use('DT-UCPH/sp:hot', hoist=globals())
```

```python
display_features = ["lex", "g_nme", "g_vbe", "g_nme",'prs_gn',"prs_nu","prs_ps"]

SP.displaySetup(extraFeatures=display_features)
MT.displaySetup(extraFeatures=display_features)

def show(w):
    SP.pretty(SP.api.L.u(w, 'verse')[0], highlights={w: 'gold'})
    bo, ch, ve = SP.api.T.sectionFromNode(w)
    mt_ref = MT.api.T.nodeFromSection((bo, int(ch), int(ve)))
    MT.pretty(mt_ref)
    
def unique_SP(feat):
    unique_SP_list = []

    MT_values = {eval(f'MT.api.F.{feat}.v({w})') for w in MT.api.F.otype.s('word')}
    for w in SP.api.F.otype.s('word'):
        if eval(f'SP.api.F.{feat}.v({w})') not in MT_values:
            unique_SP_list.append(w)
    
    return unique_SP_list

def MT_variation(feat):

    variation = []
    
    for w in SP.api.F.otype.s('word'):
        if SP.api.F.mt_feat.v(w) == 'True':
            bo, ch, ve = SP.api.T.sectionFromNode(w)
            mt_ref = MT.api.T.nodeFromSection((bo, int(ch), int(ve)))
            mt_words = MT.api.L.d(mt_ref, 'word')
            for MTw in mt_words:
                if MT.api.F.g_cons.v(MTw) == SP.api.F.g_cons.v(w) and MT.api.F.lex.v(MTw) == SP.api.F.lex.v(w):
                    if eval(f'MT.api.F.{feat}.v({MTw})') != eval(f'SP.api.F.{feat}.v({w})'):
                        variation.append(w)
    
    return variation
```
