```python
from Bio import pairwise2
from Bio.Seq import Seq
import collections
from tf.app import use
```

```python
MT = use('etcbc/bhsa')
SP = use('app:./app')
```

```python
books = ['Genesis']
```

```python
def arrangeText(feature, books, corpus):
    
    text_dict = collections.defaultdict(str)

    for w in eval(corpus + '.api.F.otype.s("word")'):
        bo, ch, ve = eval(corpus + '.api.T.sectionFromNode('+str(w)+')')
        if bo in books:
            text_dict[f'{bo} {ch} {ve}'] += f"{eval(corpus + '.api.F.' + feature + '.v(' + str(w) +')')} "
            
    return text_dict

def alignment(feature, books, corpus1='MT', corpus2='SP'):
    
    seq1 = arrangeText(feature, books, corpus1)
    seq2 = arrangeText(feature, books, corpus2)
    
    alignment_dict = {}
    for v in seq1:
        if v not in seq2:
            print(f'{v} not found in SP')

        else:
            alignment = pairwise2.align.globalxx(seq1[v], seq2[v])
            fraction = alignment[0].score/len(seq1[v])
            alignment_dict[v] = alignment, fraction
            
    return alignment_dict
```

```python
alignment_dict = alignment('g_cons', books)
```

Overall alignment

```python
from statistics import mean
```

```python
mean([n[1][1] for n in alignment_dict.items()])
```

### Alignment of lexemes

```python
alignment_lex = alignment('lex', books)
```

```python
mean([n[1][1] for n in alignment_lex.items()])
```

```python
difference = {}

for v in alignment_dict:
    if alignment_dict[v][1] != alignment_lex[v][1]:
        difference[v] = alignment_dict[v][1] - alignment_lex[v][1]
```

```python
import pandas as pd
df = pd.DataFrame([difference]).T
```
